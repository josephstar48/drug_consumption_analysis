```python
"""Converts drug use ratings in specified columns to numerical values."""

ratings_lst = cleaning_drug_df[['Alcohol', 'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms', 'Nicotine', 'Semer', 'VSA']].columns.tolist()

cleaning_drug_df_copy = cleaning_drug_df.copy()

cleaning_drug_df_v2 = convert_drug_use_ratings(cleaning_drug_df_copy, ratings_lst)
```

```python
"""Converts age ranges in Age column to their midpoint numerical values."""

age_range_lst = cleaning_drug_df_v2[['Age']].columns.tolist()

cleaning_drug_df_v2_copy = cleaning_drug_df_v2.copy()

cleaning_drug_df_v3 = convert_to_midpoint(cleaning_drug_df_v2_copy, age_range_lst)
```
